```python
import nest_asyncio
import asyncio
import aiohttp
import google.generativeai as genai
import pandas as pd


genai.configure(api_key=GOOGLE_API_KEY)

# Function to generate tokenized response
def generate_response(subject):
    prompt = (
        "Just Give only answer. You don't need to explain"
        "Tokenize the given Korean subject data to reflect the characteristics of STEM data. "
        "For example, 열역학(thermodynamics) is about 열 and 역학, 통계물리(statistical physics) consists of 통계 and 물리, "
        "전자물리 and 전자물리학 (electrophysics) are 전자 and 물리, 화학(general chemistry) is 화학, "
        "전자기학(electromagnetism) is 전자기학. "
        "유기생물, 유기생명, and 유기생물학 are about 유기 and 생물. "
        "If the data is entirely in English, such as Trends in Biotechnology, translate it into Korean and then discuss it. "
        "물리학 is the same as 물리, and 생물학 is the same as 생물. "
        f"Data: {subject}"
    )
    model = genai.GenerativeModel('gemini-1.5-pro')
    response = model.generate_content(prompt)

    try:
        # Access the first candidate's content
        candidate = response.candidates[0]
        tokens = candidate.content.parts[0].text.strip()
        return tokens
    except (KeyError, IndexError, AttributeError) as e:
        return f"Error: {str(e)} - Invalid response structure or no candidates found."

# Load the curriculum data from the Excel file
data = pd.read_excel("/content/drive/MyDrive/Curriculum Analysis Project/K-curriculum Analysis/Data/2024-1/combined_data_nonum.xlsx")
# Extract unique subjects from the curriculum
curriculum = data['교육과정'].drop_duplicates().tolist()
```

```python
len(curriculum)
```

```python
from tqdm import tqdm

response_dict = {}
response_dict_reverse = {}

# Assuming 'curriculum' is a list of subjects
for subject in tqdm(curriculum, desc="Processing subjects"):
    response = generate_response(subject)
    tokens = response.split(',')

    # Populate response_dict
    response_dict[subject] = tokens

    # Populate response_dict_reverse
    for token in tokens:
        if token in response_dict_reverse:
            response_dict_reverse[token].append(subject)
        else:
            response_dict_reverse[token] = [subject]

# Print the results
print("response_dict:", response_dict)
print("response_dict_reverse:", response_dict_reverse)
```

```python
import pandas as pd
import nest_asyncio
import google.generativeai as genai
import json
import time
from sklearn.cluster import KMeans
import numpy as np
```

```python

genai.configure(api_key=GOOGLE_API_KEY)

nest_asyncio.apply()

def cluster_courses(courses, retries=10, delay=600, column=None, cluster_idx=None):
    prompt = (
        "For the following list of subjects from the Korean Basic Natural Science Department, "
        "translate each subject into English and then assign each to a specific group based on the core topics and content they cover. "
        "This categorization requires high granularity to ensure that each cluster reflects very specific and detailed aspects of the subjects. "
        "Each group must contain courses that are not only interchangeable but also tightly focused on a narrow aspect of the sub-field, "
        "promoting an in-depth understanding and specialization. "
        "Focus intensely on the key nouns and descriptors in each course name to determine their grouping, ensuring that each group contains courses that cover the same specific topic or sub-field. "
        "For instance, subjects like 물리학 (physics), 물리학연습 (physics practice), and 물리학실험 (physics experiment) should only be grouped together if they address exactly the same concepts or laboratory techniques. "
        "On the other hand, even closely related subjects such as 고급물리학 (advanced general physics), 고급현대물리학 (advanced modern physics), and 고급통계물리학 (advanced statistical physics) should be separated into different groups if they emphasize different principles or applications within physics. "
        "Provide the result as a JSON object where each key is the subject and the value is the group number, including a detailed explanation for the clustering rationale that highlights the specific focus of each group. "
        "Note: Ensure that the JSON output does not contain comments as comments are not allowed in JSON format. "
        f"Subjects: {courses}"
    )

    for attempt in range(retries):
        try:
            model = genai.GenerativeModel('gemini-1.5-pro')
            response = model.generate_content(prompt)
            if response and response.candidates:
                message = response.candidates[0].content.parts[0].text

                # Saving Raw response
                with open(f'/content/drive/MyDrive/Curriculum Analysis Project/K-curriculum Analysis/gemini base/0717 attempt/gemini_responses_{column}_cluster{cluster_idx}.txt', 'a') as f:
                    f.write(f"Attempt {attempt + 1}: Raw response for column {column}, cluster {cluster_idx}, courses {courses}:\n{message}\n\n")

                return message  # Return a text response
            else:
                raise ValueError("No candidates found in the response.")
        except (ValueError, KeyError, IndexError, AttributeError) as e:
            print(f"Attempt {attempt + 1} failed: {str(e)}")
            if attempt < retries - 1:
                time.sleep(delay)
            else:
                return f"Error: {str(e)} - Invalid response structure or no candidates found."
        except Exception as e:
            print(f"Attempt {attempt + 1} failed: {str(e)}")
            if attempt < retries - 1:
                time.sleep(delay)
            else:
                return f"Error: {str(e)} - Request failed."
```

```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.feature_extraction.text import TfidfVectorizer

# Function to tokenize and sort courses in Korean alphabetical order without locale
def sort_korean(courses):
    def korean_key(course):
        return [ord(char) for char in course]
    return sorted(courses, key=korean_key)

def syllable_clustering(tokenized_courses_dict, courses, num_clusters=3):
    tokenized_courses = [tokenized_courses_dict.get(course, "") for course in courses]
    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(tokenized_courses).toarray()
    clustering = AgglomerativeClustering(n_clusters=num_clusters)
    clusters = clustering.fit_predict(X)

    clustered_courses = {}
    for idx, cluster_id in enumerate(clusters):
        if cluster_id not in clustered_courses:
            clustered_courses[cluster_id] = []
        clustered_courses[cluster_id].append(courses[idx])

    final_clusters = []
    for cluster_id, cluster_courses in clustered_courses.items():
        final_clusters.append(sort_korean(cluster_courses))
    return final_clusters

# Recursive clustering with tokenized data
def recursive_clustering(courses, tokenized_courses_dict, max_size=50):
    clusters = syllable_clustering(tokenized_courses_dict, courses, num_clusters=3)
    result_clusters = []
    for cluster in clusters:
        if len(cluster) >= max_size:
            result_clusters.extend(recursive_clustering(cluster, tokenized_courses_dict, max_size))
        else:
            result_clusters.append(cluster)
    return result_clusters

```

```python
def process_columns(df, column_range, tokenized_courses_dict):
    df_sample = df.iloc[:, column_range]

    for column in df_sample.columns:
        df_column = df[column].dropna()
        courses = df_column.tolist()

        if len(courses) >= 50:
            syllable_clusters = recursive_clustering(courses, tokenized_courses_dict, max_size=50)
        else:
            syllable_clusters = [courses]

        for cluster_idx, cluster in enumerate(syllable_clusters, 1):
            clusters = cluster_courses(cluster, column=column, cluster_idx=cluster_idx)

            print(f"Column: {column}, Cluster {cluster_idx}: {cluster}")
            print(f"Clustered courses: {clusters}\n")

            if isinstance(clusters, str):
                print(clusters)  # If the response is a string (like an error message), just print it.

            # Optionally save the raw cluster response to a file for later analysis or reference
            with open(f'./cluster_responses_{column}_cluster{cluster_idx}.txt', 'a') as f:
                f.write(f"Column {column}, Cluster {cluster_idx}, Courses: {courses}\n")
                f.write(f"Clustered Courses: {clusters}\n\n")

```
